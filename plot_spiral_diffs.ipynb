```python
# Plot files
import numpy as np
import matplotlib.pyplot as plt

def plot_diff(filename, f2):
    U = np.zeros((200, 200))
    U_f2 = np.zeros((200, 200))

    with open(filename) as f:
        for i in range(len(U)):
            for k in range(len(U[0])):
                line = f.readline()
                if line == '':
                    break
                U[i,k] = float(line)
    
    with open(f2) as f_f2:
        for i in range(len(U)):
            for k in range(len(U[0])):
                line = f_f2.readline()
                if line == '':
                    break
                U_f2[i,k] = float(line)

    aux = U[:,:] - U_f2[:,:]
        
    plt.figure(figsize=(8,8))
    plt.title('Difference')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(aux, cmap='plasma', vmin=-100, vmax=50)
    plt.colorbar(label='V (mV)')

    plt.show()
```

```python
# Plot files
import numpy as np
import matplotlib.pyplot as plt

def plot_file(filename, title):
    U = np.zeros((200, 200))

    with open(filename) as f:
        for i in range(len(U)):
            for k in range(len(U[0])):
                line = f.readline()
                if line == '':
                    break
                U[i,k] = float(line)

    aux = U[:,:]
        
    plt.figure(figsize=(8,8))
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(aux, cmap='plasma', vmin=-100, vmax=50)

    plt.show()
```
